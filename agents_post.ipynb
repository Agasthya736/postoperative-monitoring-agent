```python
def fetch_patient_data(patient_id):

    vitals = requests.get(
        f"http://127.0.0.1:8000/patient/{patient_id}/vitals"
    ).json()

    biomarkers = requests.get(
        f"http://127.0.0.1:8000/patient/{patient_id}/biomarkers"
    ).json()

    return {
        "vitals": vitals,
        "biomarkers": biomarkers
    }
```

```python
data = fetch_patient_data(101)
print(data)
```

```python
def analyze_patient(data):

    prompt = f"""
    You are an AI assistant monitoring a post operative patient.

    Analyze the following data and identify possible complications.

    Data:
    {data}

    Mention:
    - abnormal vitals
    - biomarker concerns
    - whether an alert is required
    """

    response = model.generate_content(prompt)

    return response.text
```

```python
def monitoring_agent(patient_id):

    data = fetch_patient_data(patient_id)

    report = analyze_patient(data)

    return report
```
